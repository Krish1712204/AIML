```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints)
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Draw a line')
plt.show()
```

```python
X=range(1,50)
Y=[value *3 for value in X]
print("Values of X:")
print(*range(1,50))

print("Values in Y(thrice of X):")
print(Y)
plt.plot(X,Y)
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
plt.title('Draw a line')
plt.show()
```

```python
# write a program to
#x axis values 
x=[1,2,3]
#y axis values
y=[2,4,1]
# plot lines and/or markers to axes
plt.plot(x,y)
#SET the x axis label of the current axis
plt.xlabel('X-axis')
#SET the y axis label of the current axis
plt.ylabel('Y-axis')
#set a title
plt.title('sample graph')
plt.show()
```
